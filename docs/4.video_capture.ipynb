### MỞ VIDEO VỚI OPENCV ###

```python
import cv2
```

```python
# đọc file video
cap = cv2.VideoCapture(r'..\data\video.mp4')

# sử dụng vòng lặp để đọc từng khung hình
while True:       # kiểm tra video còn mở không
    ret, frame = cap.read()   # trả về 2 giá trị: ret (True/False), frame (khung hình)
    if ret == True: # nếu còn khung hình
        frame = cv2.resize(frame, (640, 480))  # thay đổi kích thước khung hình
        # hiển thị khung hình
        cv2.imshow('Video', frame)
        if cv2.waitKey(25) & 0xFF == ord('q'):   #key = 25ms (tốc độ phát nhanh hay chậm), nhấn 'q' để thoát
            break
    else:
        break

cap.release()   # giải phóng bộ nhớ
cv2.destroyAllWindows()
```
