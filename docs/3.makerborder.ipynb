***ĐÁNH DẤU ĐƯỜNG VIỀN ẢNH***

```python
import cv2
import matplotlib.pyplot as plt
```

```python
#Hàm copyMakeBorder() để tạo đường viền ảnh 
'''
cv2.copyMakeBorder(	src, top, bottom, left, right, borderType[, dst[, value]]	) ->	dst
- Trong đó:
    + src: Ảnh gốc
    + top, bottom, left, right: Kích thước đường viền cần tạo ở các cạnh tương ứng
    + borderType: Kiểu đường viền (BORDER_CONSTANT, BORDER_REFLECT, BORDER_REFLECT_101/ BORDER_DEFAULT, BORDER_REPLICATE, BORDER_WRAP)
    + value: Màu sắc của đường viền nếu sử dụng kiểu BORDER_CONSTANT
- Trả về: Ảnh sau khi thêm đường viền
'''
img = cv2.imread(r'..\data\00-puppy.jpg')
cv2.resize(img, (400, 400))

# Tạo các kiểu đường viền khác nhau
bordered_img = cv2.copyMakeBorder(img, 50, 50, 100, 100, cv2.BORDER_CONSTANT, value=[0, 255, 0]) 

# Phản chiếu ảnh làm đường viền
# bordered_img = cv2.copyMakeBorder(img, 50, 50, 100, 100, cv2.BORDER_REFLECT)    

 # Nhân bản pixel cạnh làm đường viền
# bordered_img = cv2.copyMakeBorder(img, 350, 350, 100, 100, cv2.BORDER_REPLICATE) 

# Lấy phần ảnh bên kia làm đường viền
# bordered_img = cv2.copyMakeBorder(img, 50, 50, 100, 100, cv2.BORDER_WRAP)  
```
