### VIDEO CAPTURE VỚI CAMERA ###

```python
import cv2
```

```python
cap = cv2.VideoCapture(0)   # Mở camera mặc định của máy tính
while True:
    ret, frame = cap.read()  # Đọc từng khung hình từ camera
    if not ret:
        break
    cv2.imshow('Camera', frame)  # Hiển thị khung hình lên cửa sổ

    if cv2.waitKey(25) & 0xFF == ord('q'):  # Nhấn 'q' để thoát
        break

cap.release()
cv2.destroyAllWindows()
```
