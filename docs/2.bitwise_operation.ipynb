***TOÁN TỬ BITWISE TRONG ẢNH***

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread(r'..\data\00-puppy.jpg')
img2 = cv2.imread(r'..\data\sammy.jpg')

# Đưa ảnh về cùng kích thước trước khi trộn
img1 = cv2.resize(img1, (400, 400))
img2 = cv2.resize(img2, (400, 400))
```

```python
#Toán tử AND: 
# Toán tử AND thực hiện phép toán AND bitwise giữa hai ảnh. 
# Kết quả là một ảnh mới chỉ chứa các pixel mà cả hai ảnh gốc đều có giá trị khác 0 tại vị trí đó.
bitwise_and = cv2.bitwise_and(img1, img2)

#Toán tử OR:
# Toán tử OR thực hiện phép toán OR bitwise giữa hai ảnh.
# Kết quả là một ảnh mới chứa các pixel mà ít nhất một trong hai ảnh gốc có giá trị khác 0 tại vị trí đó.
bitwise_or = cv2.bitwise_or(img1, img2)

#Toán tử XOR:
# Toán tử XOR thực hiện phép toán XOR bitwise giữa hai ảnh.
# Kết quả là một ảnh mới chứa các pixel mà chỉ một trong hai ảnh gốc có giá trị khác 0 tại vị trí đó.
bitwise_xor = cv2.bitwise_xor(img1, img2)

#Toán tử NOT:
# Toán tử NOT thực hiện phép toán NOT bitwise trên một ảnh.
# Kết quả là một ảnh mới trong đó các giá trị pixel được đảo ngược (0 trở thành 255 và ngược lại).
bitwise_not_img1 = cv2.bitwise_not(img1)
```

```python
# Hiển thị kết quả bằng matplotlib
plt.figure(figsize=(10, 8))
title = ['Image 1', 'Image 2', 'Bitwise AND', 'Bitwise OR', 'Bitwise XOR', 'Bitwise NOT (Image 1)']
images = [img1, img2, bitwise_and, bitwise_or, bitwise_xor, bitwise_not_img1]
for i in range(6):
    plt.subplot(2, 3, i+1)
    plt.imshow(cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB))
    plt.title(title[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
```
