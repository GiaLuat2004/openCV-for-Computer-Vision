### X·ª¨ L√ù H√åNH TH√ÅI ·∫¢NH ###

```python
import cv2
import matplotlib.pyplot as plt
```

```python
"""
===========================================================
MORPHOLOGICAL OPERATIONS TRONG OPENCV
===========================================================

Morphological operations (c√°c ph√©p to√°n h√¨nh th√°i h·ªçc) l√† c√°c ph√©p to√°n
x·ª≠ l√Ω ·∫£nh d·ª±a tr√™n H√åNH D·∫†NG c·ªßa ƒë·ªëi t∆∞·ª£ng.

- √Åp d·ª•ng ch·ªß y·∫øu cho ·∫£nh NH·ªä PH√ÇN (0 - 255)
- D·ª±a tr√™n kernel (structuring element)
- Th∆∞·ªùng d√πng trong:
  + X√≥a nhi·ªÖu
  + L·∫•p l·ªó
  + T√°ch / n·ªëi v·∫≠t th·ªÉ
  + Ti·ªÅn x·ª≠ l√Ω tr∆∞·ªõc khi t√¨m contour
"""

# =========================================================
# 1. ƒê·ªåC ·∫¢NH
# =========================================================
# cv2.imread(path, flag)
# flag = 0  : ·∫£nh grayscale
# flag = 1  : ·∫£nh m√†u (BGR)
img = cv2.imread(r"..\data\00-puppy.jpg", 0)

if img is None:
    print("‚ùå Kh√¥ng ƒë·ªçc ƒë∆∞·ª£c ·∫£nh")
    exit()

# =========================================================
# 2. T·∫†O KERNEL (STRUCTURING ELEMENT)
# =========================================================
"""
cv2.getStructuringElement(shape, ksize)

THAM S·ªê:
- shape:
    cv2.MORPH_RECT     : kernel h√¨nh ch·ªØ nh·∫≠t
    cv2.MORPH_ELLIPSE  : kernel h√¨nh elip (tr√≤n)
    cv2.MORPH_CROSS    : kernel h√¨nh d·∫•u c·ªông

- ksize:
    (width, height) c·ªßa kernel
    -> s·ªë l·∫ª (3x3, 5x5, ...)
    -> kernel c√†ng l·ªõn, hi·ªáu ·ª©ng c√†ng m·∫°nh
"""
kernel = cv2.getStructuringElement(
    cv2.MORPH_RECT,
    (5, 5)
)

# =========================================================
# 3. GI·∫¢I TH√çCH H√ÄM cv2.morphologyEx()
# =========================================================
"""
C√ö PH√ÅP ƒê·∫¶Y ƒê·ª¶:

cv2.morphologyEx(
    src,           # ·∫¢nh ƒë·∫ßu v√†o
    op,            # Lo·∫°i ph√©p to√°n morphology
    kernel,        # Structuring element
    dst=None,      # ·∫¢nh ƒë·∫ßu ra (√≠t d√πng)
    anchor=(-1,-1),# T√¢m kernel (-1,-1 = t√¢m m·∫∑c ƒë·ªãnh)
    iterations=1,  # S·ªë l·∫ßn l·∫∑p ph√©p to√°n
    borderType,    # Ki·ªÉu x·ª≠ l√Ω bi√™n
    borderValue    # Gi√° tr·ªã bi√™n n·∫øu d√πng BORDER_CONSTANT
)

-----------------------------------------
GI·∫¢I TH√çCH C√ÅC THAM S·ªê:
-----------------------------------------

1. src:
   - ·∫¢nh ƒë·∫ßu v√†o (grayscale ho·∫∑c binary)
   - Th∆∞·ªùng l√† ·∫£nh nh·ªã ph√¢n ƒë·ªÉ ƒë·∫°t hi·ªáu qu·∫£ t·ªët nh·∫•t

2. op (QUAN TR·ªåNG NH·∫§T):
   - cv2.MORPH_ERODE     : Erosion
   - cv2.MORPH_DILATE    : Dilation
   - cv2.MORPH_OPEN      : Opening
   - cv2.MORPH_CLOSE     : Closing
   - cv2.MORPH_GRADIENT  : Gradient
   - cv2.MORPH_TOPHAT    : Top-hat
   - cv2.MORPH_BLACKHAT  : Black-hat

3. kernel:
   - Ma tr·∫≠n qu√©t ·∫£nh
   - Quy·∫øt ƒë·ªãnh m·ª©c ƒë·ªô ·∫£nh h∆∞·ªüng c·ªßa ph√©p to√°n

4. dst:
   - ·∫¢nh ƒë·∫ßu ra (OpenCV t·ª± c·∫•p ph√°t n·∫øu kh√¥ng truy·ªÅn)

5. anchor:
   - T√¢m kernel khi qu√©t
   - (-1, -1) = t√¢m kernel (m·∫∑c ƒë·ªãnh, n√™n d√πng)

6. iterations:
   - S·ªë l·∫ßn l·∫∑p ph√©p to√°n
   - iterations > 1 ‚Üí hi·ªáu ·ª©ng m·∫°nh h∆°n

7. borderType:
   - C√°ch x·ª≠ l√Ω bi√™n ·∫£nh
   - Th∆∞·ªùng d√πng: cv2.BORDER_CONSTANT

8. borderValue:
   - Gi√° tr·ªã ƒë·ªám bi√™n n·∫øu borderType = BORDER_CONSTANT
"""

# =========================================================
# 4. EROSION (PH√âP CO)
# =========================================================
eroded = cv2.morphologyEx(
    src=img,
    op=cv2.MORPH_ERODE,
    kernel=kernel,
    iterations=1
)

# =========================================================
# 5. DILATION (PH√âP GI√ÉN)
# =========================================================
dilated = cv2.morphologyEx(
    src=img,
    op=cv2.MORPH_DILATE,
    kernel=kernel,
    iterations=1
)

# =========================================================
# 6. OPENING (EROSION ‚Üí DILATION)
# =========================================================
opening = cv2.morphologyEx(
    src=img,
    op=cv2.MORPH_OPEN,
    kernel=kernel,
    iterations=1
)

# =========================================================
# 7. CLOSING (DILATION ‚Üí EROSION)
# =========================================================
closing = cv2.morphologyEx(
    src=img,
    op=cv2.MORPH_CLOSE,
    kernel=kernel,
    iterations=1
)

# =========================================================
# 8. MORPHOLOGICAL GRADIENT (BI√äN)
# =========================================================
gradient = cv2.morphologyEx(
    src=img,
    op=cv2.MORPH_GRADIENT,
    kernel=kernel
)

# =========================================================
# 9. TOP-HAT
# =========================================================
tophat = cv2.morphologyEx(
    src=img,
    op=cv2.MORPH_TOPHAT,
    kernel=kernel
)

# =========================================================
# 10. BLACK-HAT
# =========================================================
blackhat = cv2.morphologyEx(
    src=img,
    op=cv2.MORPH_BLACKHAT,
    kernel=kernel
)

# =========================================================
# 11. HI·ªÇN TH·ªä K·∫æT QU·∫¢
# =========================================================
titles = [
    "Original", "Eroded", "Dilated", "Opening",
    "Closing", "Gradient", "Top-hat", "Black-hat"
]
images = [
    img, eroded, dilated, opening, closing, gradient, tophat, blackhat
]
plt.figure(figsize=(12, 8))
for i in range(len(images)):
    plt.subplot(2, 4, i + 1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')
plt.tight_layout()
plt.show()

"""
===========================================================
T√ìM T·∫ÆT NHANH
===========================================================

- Erosion    : co v·∫≠t th·ªÉ, x√≥a nhi·ªÖu tr·∫Øng
- Dilation   : gi√£n v·∫≠t th·ªÉ, l·∫•p l·ªó
- Opening    : x√≥a nhi·ªÖu nh·ªè
- Closing    : l·∫•p l·ªó nh·ªè
- Gradient   : ph√°t hi·ªán bi√™n
- Top-hat    : chi ti·∫øt s√°ng nh·ªè
- Black-hat  : chi ti·∫øt t·ªëi nh·ªè

üëâ Quan tr·ªçng nh·∫•t khi d√πng morphology:
- op
- kernel
- iterations
"""
```
