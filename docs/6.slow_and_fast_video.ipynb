### Tốc độ phát nhanh và chậm đối với video ###

```python
import cv2
```

```python
cap = cv2.VideoCapture(r'..\data\video.mp4')

while True:
    ret, frame = cap.read()
    if not ret:
        break
    cv2.imshow('Video', frame)
    if cv2.waitKey(50) & 0xFF == ord('q'):  
        # giải thích về key:
        # waitKey(delay) trả về mã phím đã nhấn trong khoảng delay milliseconds
        # & 0xFF để lấy 8 bit thấp nhất, phù hợp với ASCII

        break
cap.release()
cv2.destroyAllWindows()
```
