```python
import cv2
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('00-puppy.jpg')
img2 = cv2.imread('sammy.jpg')

# Đưa ảnh về cùng kích thước trước khi trộn
img1 = cv2.resize(img1, (400, 400))
img2 = cv2.resize(img2, (400, 400))

# hàm addWeighted để trộn hai ảnh
'''
tham số:
- img1: ảnh thứ nhất
- alpha: hệ số trọng số của ảnh thứ nhất (theo tỷ lệ phần trăm biểu thị mức độ sáng của ảnh)
- img2: ảnh thứ hai
- beta: hệ số trọng số của ảnh thứ hai (theo tỷ lệ phần trăm biểu thị mức độ sáng của ảnh)
- gamma: hằng số được thêm vào mỗi pixel
'''
new_img = cv2.addWeighted(img1, 0.7, img2, 1.2, 0)

#hàm subtract để trừ hai ảnh (trừ pixel của ảnh thứ hai khỏi pixel của ảnh thứ nhất)
'''
tham số:
- img1: ảnh thứ nhất
- img2: ảnh thứ hai
'''
# new_img2 = cv2.subtract(img1, img2)
plt.imshow(cv2.cvtColor(new_img, cv2.COLOR_BGR2RGB))
# plt.imshow(cv2.cvtColor(new_img2, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
